## II. Random Forest

**Split the data into training and test sets**

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=666)
```

**Define the parameter grid for Grid Search**

```python
param_grid = {
    'max_depth': [3, 5, 7, 10, 15, 20, None],  # Different depths of the decision trees
    'min_samples_split': [2, 5, 10, 15, 20]    # Minimum number of samples required to split an internal node
}
```

**Initialize the RandomForestClassifier**

```python
rfc = RFC(random_state=666)
```

**Perform Grid Search with Cross-Validation to find the best parameters**

```python
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)
```

**Print the best parameters and the best score from Grid Search**

```python
print(f"Best parameters: {grid_search.best_params_}")
```

```python
print(f"Best score: {grid_search.best_score_:.4f}")
```

**Train the RandomForestClassifier with the best parameters**

```python
best_rfc = grid_search.best_estimator_
y_pred = best_rfc.predict(X_test)
```

**Evaluate the model**

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy:.4f}")
```

```python
print("Classification report:\n", classification_report(y_test, y_pred))
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

**Feature Importance Visualization**

```python
plt.figure(figsize=(12, 8))
importances = best_rfc.feature_importances_
indices = np.argsort(importances)[::-1]
sns.barplot(x=importances[indices], y=X_train.columns[indices])
plt.title('Feature Importances')
plt.show()
```

**Predict the classification for the test data**

```python
test_data = pd.read_csv("test.csv")
X_true_test = test_data.drop(['query_id', 'url_id', 'id'], axis=1, errors='ignore')
X_true_test_scaled = scaler.transform(X_true_test)
predictions = best_rfc.predict(X_true_test_scaled)
results = pd.DataFrame({
    'prediction': predictions,
    'id': test_data['id']
})

results.to_csv("RandomForest_hyper.tuned_predictions.csv", index=False)
```

## III. Boosting

**Split the data into training and test sets**

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=666)
```

**Define the parameter grid for Grid Search**

```python
param_grid = {
    'n_estimators': [50, 100, 150, 200],       # Number of boosting stages to be run
    'learning_rate': [0.01, 0.05, 0.1, 0.2],  # Step size for updating weights
    'max_depth': [3, 5, 7, 10],                # Maximum depth of the individual trees
    'min_samples_split': [2, 5, 10],           # Minimum number of samples required to split an internal node
    'subsample': [0.8, 0.9, 1.0],             # Fraction of samples used for fitting individual trees
    'max_features': [None, 'sqrt', 'log2']    # Number of features to consider for the best split
}
```

**Initialize GridSearchCV with GradientBoostingClassifier**

```python
grid_search = GridSearchCV(estimator=GBC(random_state=666),
                           param_grid=param_grid,
                           cv=5,                           # Number of cross-validation folds
                           scoring='accuracy',             # Scoring metric
                           n_jobs=-1)                       # Use all available CPUs
```

**Fit GridSearchCV to the training data**

```python
grid_search.fit(X_train, y_train)
```

**Print the best parameters and the best score**

```python
print(f"Best parameters: {grid_search.best_params_}")
```

```python
print(f"Best score: {grid_search.best_score_:.4f}")
```

**Train the GradientBoostingClassifier with the best parameters**

```python
best_gbc = grid_search.best_estimator_
y_pred = best_gbc.predict(X_test)
```

**Evaluate the model**

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy:.4f}")
```

```python
print("Classification report:\n", classification_report(y_test, y_pred))
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

**Feature Importance Visualization**

```python
plt.figure(figsize=(12, 8))
importances = best_gbc.feature_importances_
indices = np.argsort(importances)[::-1]
sns.barplot(x=importances[indices], y=X_train.columns[indices])
plt.title('Feature Importances')
plt.show()
```

**Predict the classification for the test data**

```python
test_data = pd.read_csv("test.csv")
X_true_test = test_data.drop(['query_id', 'url_id', 'id'], axis=1, errors='ignore')
X_true_test_scaled = scaler.transform(X_true_test)
predictions = best_gbc.predict(X_true_test_scaled)
results = pd.DataFrame({
    'prediction': predictions,
    'id': test_data['id']
})

results.to_csv("Boosting_hyper.tuned_predictions.csv", index=False)
```
